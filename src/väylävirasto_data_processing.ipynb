```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
stations = pd.read_csv('/home/daniel/projects/rails/data/RailTemperatures2023StationListv2.csv', sep = ';', encoding='Latin-1')
stations
```

```python
df= pd.read_csv(r'/home/daniel/projects/rails/data/RailTemperatures2023v2.csv', sep = ';')
df
```

```python
dfs = df.merge(stations,on = 'StationId', how = 'left')
dfs
```

```python
dfs.columns
```

```python
cols = ["Station", "StationId", "Lat", "Lon", "X", "Y", "Timestamp", "TAir", "TRail"]
df = dfs[cols]
df
```

```python
df = df.rename(columns={"StationId": "station_id", "Lat": "lat", "Lon": "lon", "Station": "station"})
```

```python
df.isnull().values.any()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# function to round the minutes to the nearest hour as the ECMWF data is for every hour
from datetime import datetime, timedelta
df['Timestamp'] = pd.to_datetime(df['Timestamp'], dayfirst=True)


def hour_rounder(t):
    # Rounds to nearest hour by adding a timedelta hour if minute >= 30
    return (t.replace(second=0, microsecond=0, minute=0, hour=t.hour)
               +timedelta(hours=t.minute//30))
df['Timestamp'] = df['Timestamp'].apply(hour_rounder)
```

```python
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates(keep='first')
```

```python
df.shape
```

```python
df.dtypes
```

```python
cols = df.select_dtypes(include=[np.float64]).columns
cols
```

```python
df[cols] = df[cols].astype(np.float32)
df.dtypes
```

```python
cols = df.select_dtypes(include=[np.int64]).columns
cols
```

```python
df[cols] = df[cols].astype(np.int32)
df.dtypes
```

```python
# remove rail temperature greater than 50 degree Celsius
df =  df.loc[df.TRail <= 50]
df
```

```python
df.to_csv('/home/daniel/projects/rails/data/RailTemperatures_with_stations_2023v2.csv', encoding = 'Latin-1', index = False)
```
